```python
import json
model_results = json.load(open('all_results.json','r',encoding='utf-8'))
model_results
```

```python
# Get all keys of model names from results
model_names = list(model_results.keys())
model_names
```

```python
list(model_results['ha'].keys())
```

```python
lowest_mae = {}
for model_name in model_names:
    lowest_mae[model_name] = {}
    for key in model_results[model_name].keys():
        lowest_mae[model_name][key] = min(model_results[model_name][key])
```

```python
lowest_mae_overall = {}
for model_name in model_names:
    lowest_mae_overall[model_name] = min(lowest_mae[model_name].values())
```

```python
lowest_mae_overall = {}
for model_name in model_names:
    lowest_mae_overall[model_name] = min(lowest_mae[model_name], key=lambda x:lowest_mae[model_name][x])
lowest_mae_overall
```
