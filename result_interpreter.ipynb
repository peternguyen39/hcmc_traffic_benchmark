```python
# Get all keys of model names from results
model_names = list(model_results.keys())
model_names
```

```python
lowest_mae = {}
for model_name in model_names:
    lowest_mae[model_name] = {}
    for key in model_results[model_name].keys():
        lowest_mae[model_name][key] = min(model_results[model_name][key])
```

```python
lowest_mae_overall = {}
for model_name in model_names:
    lowest_mae_overall[model_name] = min(lowest_mae[model_name].values())
```

```python
lowest_mae_overall = {}
for model_name in model_names:
    # lowest_mae_overall[model_name] = min(lowest_mae[model_name], key=lambda x:lowest_mae[model_name][x])
    # lowest_mae_overall[model_name] = min(lowest_mae[model_name], key=lowest_mae[model_name].get)
    lowest_mae_overall[model_name] = min(lowest_mae[model_name].values())
lowest_mae_overall
```

```python
lowest_mae
```

```python
lowest_mae_per_day = {}
for model_name in model_names:
    lowest_mae_per_day[model_name] = {}
    for key in model_results[model_name].keys():
        date = key.split("'")[1].strip()
        if date not in lowest_mae_per_day[model_name]:
            lowest_mae_per_day[model_name][date] = lowest_mae[model_name][key]
        else:
            lowest_mae_per_day[model_name][date] = min(lowest_mae_per_day[model_name][date], lowest_mae[model_name][key])
lowest_mae_per_day
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
for model_name in model_names:
    plt.plot(list(lowest_mae_per_day[model_name].values()), label=model_name)
plt.xlabel('Day')
plt.ylabel('MAE')

plt.legend()
plt.show()
```

```python
with open('./data/hcmc-traffic-data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
testing_dates=['2022-05-05', '2022-05-05_1', '2022-05-06', '2022-05-07', '2022-05-07_1', '2022-05-08', '2022-05-09']
data.keys()
```

```python
sensor=0
time_data = {k:v for k,v in sorted(data[testing_dates[0]][str(sensor)].items(),key=lambda x:x[0].split('_')[1].split('.')[0])}
time_data
```

```python
import datetime
graph_data = []
for k in time_data:
    current_time = datetime.datetime.fromtimestamp(int(k.split('_')[1].split('.')[0])/1000)
    count = time_data[k]['count']
    graph_data.append((current_time, count))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame(graph_data, columns=['time', 'count'])
df['time']= pd.to_datetime(df['time']).dt.strftime('%H:%M')
ax = sns.lineplot(x='time', y='count', data=df)
ax.set(xlabel='Time', ylabel='Count')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
for ind, label in enumerate(ax.get_xticklabels()):
    if ind % 10 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```
